##### Q1.) Now your friend has followed a lot of different food bloggers, he needs to analyse the habits of these bloggers.
##### i.) From the list of instagram handles you obtained when you searched ‘food’ in previous project. Open the first 10 handles and find the top 5 which have the highest number of followers
##### ii.) Now Find the number of posts these handles have done in the previous 3 days.
##### iii.) Depict this information using a suitable graph.


```python
# Creating a list of the first 10 results on searching "food" in the search baar

# Opening instagram.
driver.get('https://instagram.com/')

# Getting the search box clearing it and typing "food".
search_box = driver.find_element_by_xpath("//input[@placeholder='Search']")
search_box.clear()
search_box.send_keys("food")

# Getting the uderids visible just after typing "food".
user_ids = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='fuqBx']//a")))


# Storing user_handles in a list .
top10 = []
for i in user_ids[:10]:
    html = i.get_attribute('outerHTML')
    data = BeautifulSoup(html,'html.parser').a['href'].split('/')[1]
    if data!='explore':
        top10.append(data)
```

```python
# Function to seacrh and open a profile (From instagram project-1) 
def search_and_open(handle):
    driver.get('https://instagram.com')
    search_input = driver.find_element_by_xpath("//input[@placeholder='Search']")
    search_input.clear()
    search_input.send_keys(handle)

    user_id = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//div[@class="fuqBx"]//a')))
    user_id.click()
    time.sleep(5)
    

# This function is used to get the follower count for any user handle.(The account must public.)
def get_follower_count(user_handle): 
    time.sleep(1)
    search_and_open(user_handle)
    tag = driver.find_element_by_partial_link_text("followers")
    data = BeautifulSoup(tag.get_attribute('outerHTML'),'html.parser').find('span').get('title')
    follower_count = int(data.replace(',',''))
    return follower_count
    
```

```python
# Top 5 accounts and their number of followers.

top5dict = {}

for i in top10 :
    no_of_followers = get_follower_count(i)
    top5dict[i] = no_of_followers
    
top5dict = dict(sorted(top5dict.items(), key=lambda x: x[1], reverse=True)[:5])
```

```python
# Top 5 accounts based on number of followers on searching food are :-

top5handles = top5dict.keys()
top5followers = top5dict.values()

df = pd.DataFrame(list(zip(top5handles, top5followers)), columns =['User handle', 'Number of followers']) 

display(df)

# Plotting a bar graph

plt.bar(top5handles , top5followers)
plt.xticks(rotation = 90)
plt.xlabel('User handles',labelpad=20)
plt.ylabel('Number of followers',labelpad=20)
plt.title('Top 5 accounts in searching "food" based on number of followers',pad=20)
plt.show()
```

```python
# Number of posts done in the last 3 days.

# This function takes in the user handle to get the count of post the user has done in last three days. It opens the profile 
# and then extracts the date and then checks the days difference from current date.
# If the difference is less or equal to 3 then it increments the no_of_posts variable. If difference is greater than 3 it breaks
# and returns the no_of_posts value and append this value to post_counts list.Try and except block have been used to handle the
# case in which the user has not posted anything.

def get_3_days_posts(user_handle):
    
    search_and_open(user_handle)
    post = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//div[@class='v1Nh3 kIKUG  _bz0w']")))
    post.click()
    no_of_posts = 0
    try:
        while True:
            time_of_post = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//a[@class='c-Yi7']/time")))
            date_of_post = BeautifulSoup(time_of_post.get_attribute('outerHTML'),'html.parser').find('time').get('title')
            date_of_post = date_of_post.replace(',', '')
            post_date = datetime.strptime(date_of_post, '%b %d %Y').date()
            current_date = datetime.today().strftime('%Y-%m-%d')
            current_date = datetime.strptime(current_date, '%Y-%m-%d').date()
            difference_in_date = (current_date-post_date).days
            next_post = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//div/a[@class=' _65Bje  coreSpriteRightPaginationArrow']")))
            next_post.click()
            if difference_in_date>3:
                break
                
            no_of_posts+=1
            time.sleep(2)
    except:
        return no_of_posts
        
    return no_of_posts;
```

```python
# Forming a dictionary with handle as keys and no of posts in last three days a values.

user_to_posts = {}

for i in top5handles:
    user_to_posts[i] = get_3_days_posts(i)
    

user_to_posts = dict(sorted(user_to_posts.items(), key=lambda x: x[1], reverse=True))

users = user_to_posts.keys()
no_of_posts = user_to_posts.values()


dataf = pd.DataFrame(list(zip(users, no_of_posts)), columns =['User handle', 'Number of posts in last 3 days']) 
display(dataf)

# Plotting a bar graph

plt.bar(users , no_of_posts)
plt.xticks(rotation = 90)
plt.xlabel('User handles',labelpad=20)
plt.ylabel('Number of posts',labelpad=20)
plt.title("Top 5 account's posts in last 3 days",pad=20)
plt.show()
```

##### Q2.) Your friend also needs a list of hashtags that he should use in his posts.
##### Open the 5 handles you obtained in the last question, and scrape the content of the first 10 posts of each handle.
##### Prepare a list of all words used in all the scraped posts and calculate the frequency of each word.
##### Create a csv file with two columns : the word and its frequency
##### Now, find the hashtags that were most popular among these bloggers
##### Plot a Pie Chart of the top 5 hashtags obtained and the number of times they were used by these bloggers in the scraped posts.


```python
# This function searches for the given user handle and oepns it. The checks recent 10 posts of a handle.It counts the likes on
# each of these posts. If its a normal post like can be directly extracted but if its a video then first the video views is
# and then the likes are extracted.The like are then accordingly added to get total likes for the handle for the last 10 posts.
# This returns content of posts , likes and numbre of posts.

def getPostDetails(handle):
    
    search_and_open(handle)
    
    post = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//div[@class='v1Nh3 kIKUG  _bz0w']")))
    post.click()
    no_of_posts = 1
    total_likes = 0
    post_for_like = 0
    texts = ""
    try:
        while True:
            content = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//div[@class='C4VMK']/span")))
            text = BeautifulSoup(content.get_attribute('outerHTML'),'html.parser').find('span').text
            texts = texts + str(" ") + str(text)
            like_data = driver.find_element_by_xpath("//section[@class='EDfFK ygqzn']")
            likes = BeautifulSoup(like_data.get_attribute('outerHTML'),'html.parser').find('span').text.replace(',','')
            if "views" not in likes:
                total_likes += int(likes)
                post_for_like+=1
            else:
                button_like = driver.find_element_by_xpath('//span[contains(.," views")]')
                driver.execute_script("arguments[0].click();", button_like)
                like_data = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//section[@class='EDfFK ygqzn']//div[contains(.,' likes')]")))
                like_html_data = BeautifulSoup(like_data.get_attribute('outerHTML'),'html.parser')
                likes = int(like_html_data.find(class_ = 'vJRqr').find('span').text.replace(",",''))
                total_likes +=likes
                post_for_like+=1


            next_post = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//div/a[@class=' _65Bje  coreSpriteRightPaginationArrow']")))
            driver.execute_script("arguments[0].click();", next_post)

            if no_of_posts==10:
                break

            no_of_posts+=1
            time.sleep(2)
    except:
        print("Failed")
        return [texts,total_likes,post_for_like]
        
    return [texts,total_likes,post_for_like]        

# Some Special characters can be removed as are unncessary.
unwanted_chars = "[-/\"/=/+/“/”/*/-/👀/📷/😂/❤️/💪/🤦/🙄/😩/🤦🏻/🤤/🌍/🍜/🇫🇷/👉/😱/🍕/🙌/🍞/🍩/🍫/🍌/🍓/☕️/🍻/✈️/🍰/🍣/🌎/🍨/🍔/🍫/♂️/♀️/🍗/🥩/🍖/🌶/😋/👨‍/😮/😍/🍳/@/,/?/🔥/👏/😉/)/🥓/(/|/0-9/./:/!/💕/🌮/>/</🍴/🍅/🧀/🎉/🔒]"


# This function takes three arguments content from last 10 posts ,word_dictionary and hastag_dictionary.
# We first split the string to getwords list. After this we iterate over all the  words
# and remove special characters in the words. If the word is not empty then we check if the first character is "#" or not and
# accordingly store in the hastag_dictionary or the word_dictionary. 

def get_word_frequency(content,words_dictionary,hastag_dictionary):
    words = content.split(" ")
    for word in words:
        word = re.sub(unwanted_chars, "", word)
        if word!="":
            if word[0]=="#":
                hastag_dictionary[word] = hastag_dictionary.get(word,0)+1
                continue
            words_dictionary[word] = words_dictionary.get(word,0)+1
            
```

```python
post_contents = []

for i in list(top5handles):
    post_content = getPostDetails(i)
    post_contents.append(post_content)
    time.sleep(3)

# Making a list of all words used in all the scraped posts and finding the frequency of each word.    
words_dictionary = {}

# Making a list of all #hashtag used in all the scraped posts and finding the frequency of each hastag.    
hastag_dictionary = {}

for content in post_contents:
    get_word_frequency(content[0],words_dictionary,hastag_dictionary)
    

words = [ key for key,value in words_dictionary.items()]
word_frequency = [ value for key,value in words_dictionary.items()]

# Creating a csv file with two columns - the word and their corresponding frequency.
columns = {"Word":words,"Frequency":word_frequency}
words_df = pd.DataFrame(columns)
words_df.to_csv(r'WordFrequency.csv', index=False)

# Finding the hashtags that were most popular among these bloggers
hastag_with_frequency = [[value,key] for key,value in hastag_dictionary.items()]
hastag_with_frequency.sort(reverse=True)

# Plotting a Pie Chart of the top 5 hashtags obtained and the number of times they were used by these bloggers in their last 10 posts.

top5_hastags = [ i[1] for i in hastag_with_frequency[:5]]
hastag_counts = [ i[0] for i in hastag_with_frequency[:5]]
plt.pie(hastag_counts, labels=top5_hastags , autopct="%.2f")
plt.show()
    
```

##### Q3.) You need to also calculate average followers : likes ratio for the obtained handles.
##### Followers : Likes ratio is calculated as follows:
##### Find out the likes of the top 10 posts of the 5 handles obtained earlier.
##### Calculate the average likes for a handle.
##### Divide the average likes obtained from the number of followers of the handle to get the average followers:like ratio of each handle.
##### Create a bar graph to depict the above obtained information.


```python
# Here we calculate the  followers to average like ratios.
# We have first found out the total likes the top 5 handles got in their last 10 posts. Then we find their average likes.This
# gives us an idea about number of likes they get on each post. 
# The we calculate the follower to average likes ratio. We have first found out the number of followers they have and then
# divided it by the average likes. We have plotted a bar graph for the same. Now using this we can get an estimate of what  
# proportion of followers of a handle like their posts.


top5handles_list = list(top5handles)

# Likes of the top 10 posts of the 5 handles obtained earlier.

print("\nHandles with most likes on last 10 Posts\n")
for i in range(5):
    print(list(top5handles)[i],post_contents[i][1])
    
# Average likes for a handle.
print("\nAverage Likes\n")
average_likes_list = []
for i in range(5):
    average_likes = post_contents[i][1]/post_contents[i][2]
    average_likes_list.append(average_likes)
    print(top5handles_list[i],format(average_likes,".0f"))

# Average Follower to like ratio
print("\nAverage follower to like ratio\n")
follower_to_avglikes_ratio = []
for i in range(5):
    followers = get_follower_count(top5handles_list[i])
    ratio = int(followers/average_likes_list[i])
    follower_to_avglikes_ratio.append(ratio)
    print(top5handles_list[i],"\t",followers," : ",format(average_likes_list[i],".0f"))

    
# Create a bar graph to depict the above obtained information.
plt.figure(figsize=(20,10))
plt.bar(top5handles_list,follower_to_avglikes_ratio)
plt.title('Average followers : Likes ratio ')
plt.grid()
plt.show()
```
